import cac thu vien va load data

```python
import pandas as pd
import matplotlib.pyplot as plt
from pymongo import MongoClient
import numpy as np
import seaborn as sns

df=pd.read_csv('./laptopData.csv')

df
```

xem xet cac row co it nhat 1 gia tri null

```python
null_row=df.isnull()

null_row=df[null_row.any(axis=1)]

null_row
```

Ta co the thay cac ban ghi co rat nhieu gia tri null nen ta se xoa het cac row co it nhat 1 gia tri null. 

```python
df=df.dropna() # loai bo cac ban ghi co it nhat 1 cot la null

df # hien thi data 
```

Ta se drop truong Unname: 0 vi no chinh la truong STT

```python
print(df.columns)
df.drop('Unnamed: 0', axis=1, inplace=True)
```

Clean truong Inches 

Xem xet cac gia tri trong truong Inches va ve bieu do de quan sat de hon

```python
# Tim cac gia tri doc nhat xuat hien trong truong Inches
unique_value=set(df['Inches']) # Dung set de loc cac gia tri trung

# In ra man hinh cac gia tri
print(unique_value)
```

Ta co the thay truong Inches co ca gia tri khong hop le nhu '?' va cac outlier.

Loai bo cac gia tri khong hop le va chuyen truong Inches sang dang so 

```python
# chuyen cac gia tri trong truong Inches thanh so, neu khong chuyen 
# duoc thi se chuyen thanh null
df.loc[:,'Inches']=pd.to_numeric(df['Inches'], errors='coerce')

# quan sat cac ban ghi bi null
all_null_row=df[df.isnull().any(axis=1)]
all_null_row
```

```python
# Gen gia tri inches theo 5 gia tri co tan so cao nhat
top5_inches=df['Inches'].value_counts().index[:5].to_list()
# dien tat cac cac gia tri null trong truong inches bang cach chon ngay nhien 1 trong 5 
df.loc[:,'Inches']=df['Inches'].apply(lambda x: np.random.choice(top5_inches) if pd.isnull(x) else x)
# kiem tra xem trong tuong inches con gia tri null nao khong neu khong thi tra ve false
print(df['Inches'].isnull().any())
```

Ve bieu do cho truong inches de de quan sat du lieu

```python
# ve bieu do cot 
plt.figure(figsize=(10,6))
plt.hist(df['Inches'], bins=20, color='skyblue', edgecolor= 'black')
plt.xlabel('Inches')
plt.ylabel('Frequency')
plt.title('Bieu do cot theo Inches')
plt.show()

#Ve bieu do tron
all_value=df['Inches'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(all_value, labels=all_value.index, autopct='%1.1f%%', startangle=140)
plt.title('Bieu do tron cho truong Inches')
plt.show()
```

Su dung IQR de loai bo outlier

```python
Q1=df['Inches'].quantile(0.25)
Q3=df['Inches'].quantile(0.75)
IQR= Q3-Q1
# tim ra cac muc voi IQR
lower_bound=Q1 - 1.5*IQR
upper_bound= Q3 + 1.5*IQR
print( upper_bound, '  ', lower_bound)

df.loc[:,'Inches']=df['Inches'].where((df['Inches']>=lower_bound) & (df['Inches']<=upper_bound))

all_null_row=df[df.isnull().any(axis=1)]
all_null_row
print('len =',len(all_null_row))
```

Generate cac gia tri cho cac outlier

```python
top5_inches=df['Inches'].value_counts().nlargest(5) # 5 gia tri pho bien nhat

prob=top5_inches/top5_inches.sum() # tinh xac xuat cho tung gia tri

# chon cac gia tri theo phan phoi da cho truoc
df.loc[:,'Inches']=df['Inches'].apply(lambda x: np.random.choice(top5_inches.index, p=prob) if pd.isnull(x) else x)

# ve bieu do cot 
plt.figure(figsize=(10,6))
plt.hist(df['Inches'], bins=20, color='skyblue', edgecolor= 'black')
plt.xlabel('Inches')
plt.ylabel('Frequency')
plt.title('Bieu do cot theo Inches')
plt.show()

#Ve bieu do tron
all_value=df['Inches'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(all_value, labels=all_value.index, autopct='%1.1f%%', startangle=140)
plt.title('Bieu do tron cho truong Inches')
plt.show()
```

Kiem tra cac ban ghi khong phai la so

```python
# Lọc các bản ghi không phải số
non_numeric_records = df[~df['Inches'].apply(lambda x: isinstance(x, (int, float)))]

# In các bản ghi không phải số
print(non_numeric_records)
```

Clean truong ScreenResolution

Co hai thu rat quan trong cua mot cai man hinh trong dataset nay la do phan giai man hinh va tam nen va cai man hinh do co cam ung duoc khong

Kiem tra xem truong ScreenResolution co cac tam nem man hinh gi

```python
# Kiem tra xem trong data co cac loai ma hinh gi (khong phan biet hoa, thuong)
oled=df[df['ScreenResolution'].str.contains('oled', case=False, na=False)]
print('oled  len=',len(oled))
print(oled)
# Tam nen IPS
ips=df[df['ScreenResolution'].str.contains('ips', case=False, na=False)]
print('ips  len=', len(ips))
print(ips)
# Tam nen TN
tn=df[df['ScreenResolution'].str.contains('tn', case=False, na=False)]
print('tn len=', len(tn))
print(tn)
```

Ta nhan thay chi co tam nen IPS duoc liet ke trong dataset nay

Lay do phan gia man hinh tu data

```python
# loai lay do phan giai man hinh bang regex
df['Resolution']=df['ScreenResolution'].str.extract(r'(\d{3,4}+x+\d{3,4})')
# in data
df
```

tao cac cot dummy_value de phan loai do phan giai man hinh, cung nhu tam nen

```python
# tao cot dummy_res de phan loai do phan giai
df['dummy_res']=df['Resolution'].apply(lambda x: 1 if x[0]=='1' else
                                                2 if x[0]=='2' else
                                                3 if x[0]=='3' else np.nan())
# kiem tra xem co ban ghi nao khong the chuyen thanh dummy value duoc khong
is_null=df['dummy_res'].isnull().any() 
is_null # ta thay tat ca deu tot
```

Ve bieu do de the hien do phan giai man hinh

```python
# Ve bieu do tron
plt.figure(figsize=(6,6))
plt.pie(df['dummy_res'].value_counts().values, labels=df['dummy_res'].value_counts().index, autopct='%1.1f%%', startangle=124)
plt.title('Bieu do the hien phan bo cua dummy_res')
plt.show()

plt.figure(figsize=(6,6))
plt.pie(df['dummy_panel'].value_counts().values, labels=df['dummy_panel'].value_counts().index, autopct='%1.1f%%', startangle=124)
plt.title('Bieu do the hien phan bo cua dummy_panel')
plt.show()

plt.figure(figsize=(6,6))
plt.pie(df['is_touchable'].value_counts().values, labels=df['is_touchable'].value_counts().index, autopct='%1.1f%%', startangle=124)
plt.title('Bieu do the hien phan bo cua man hinh cam ung')
plt.show()
```

Tao cot dummy_panel cho man hinh IPS 

```python
# neu la man hinh IPS thi dummy_panel se la 1 nguoc lai la 0
df['dummy_panel']=df['ScreenResolution'].str.contains('ips', case=False, na=False).apply(lambda x: 1 if x else 0)
# kiam tra xem co gia tri nao khong chuyn duoc
is_null=df['dummy_panel'].isnull().any()
is_null # tra ve False tuc la tot
```

Tao cot is_touchable de danh dau la man hinh cam ung

```python
df['is_touchable']=df['ScreenResolution'].apply(lambda x: 1 if 'touchscreen' in x.lower() else 0)
df
```

Clean truong Weight

Loai bo 'kg' khoi truong Weight

```python
df.loc[:,'Weight']=df['Weight'].str.removesuffix('kg')
```

Chuyen truong Weight thanh so va generate gia tri cho cai ban ghi null

```python
# chuyen Truong Weight thanh so neu khong chuyen dc thi thanh null
df.loc[:,'Weight']=pd.to_numeric(df['Weight'], errors='coerce')
# Kiem tra xem co ban ghi nao khong the chuyen thanh so 
print(df['Weight'].isnull().any())

# Generate gia tri cho cac ban ghi null

#Lay top 5 gia tri co tan so cao nhat
top5_Weight=df['Weight'].value_counts().nlargest(5)
prob=top5_Weight / top5_Weight.sum() # lay xac xuat tung gia tri
df.loc[:,'Weight']=df['Weight'].apply(
    lambda x: np.random.choice(top5_Weight.index,p=prob) if pd.isnull(x) else x)

print(df['Weight'].isnull().any()) # Moi thu deu on
```

Ve bieu do de de quan sat truong Weight de hon 

```python
# Ve bieu do cot
plt.figure(figsize=(10,6))
plt.hist(df['Weight'], bins=20,color='skyblue', edgecolor='black')
plt.xlabel( 'Weight')
plt.ylabel('Frequency')
plt.title('Bieu do the hien phan bo cua Weight')
plt.show()
```

Ta thay truong Weigh van con mot so outlier, ta se su dung IQR de loai bo no roi se generate lai no.

```python
Q1=df['Weight'].quantile(0.25)
Q3=df['Weight'].quantile(0.75)
IQR= Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound, upper_bound)

df.loc[(df['Weight']<lower_bound) | (df['Weight']>upper_bound) ,'Weight']=np.nan

#Lay top 5 gia tri co tan so cao nhat
top5_Weight=df['Weight'].value_counts().nlargest(5)
prob=top5_Weight / top5_Weight.sum() # lay xac xuat tung gia tri
# generate gia tri theo phan bo 
df.loc[:,'Weight']=df['Weight'].apply(lambda x: np.random.choice(top5_Weight.index,p=prob) if pd.isnull(x) else x)
# kiem tra con gia tri null nao khong
print(df['Weight'].isnull().any()) # Moi thu deu on
```

Ve bieu do

```python
# Ve bieu do cot
plt.figure(figsize=(10,6))
plt.hist(df['Weight'], bins=20,color='skyblue', edgecolor='black')
plt.xlabel( 'Weight')
plt.ylabel('Frequency')
plt.title('Bieu do the hien phan bo cua Weight')
plt.show()

all_value=df['Weight'].value_counts()
# Ve bieu do tron
plt.figure(figsize=(8,8))
plt.pie(all_value, labels=all_value.index, autopct='%1.1f%%', startangle=90)
plt.title('Phan bo cau weight')
plt.show()
```
