Clean truong Inches 

Xem xet cac gia tri trong truong Inches va ve bieu do de quan sat de hon

```python
# Tim cac gia tri doc nhat xuat hien trong truong Inches
unique_value=set(df['Inches']) # Dung set de loc cac gia tri trung

# In ra man hinh cac gia tri
print(unique_value)
```

Ta co the thay truong Inches co ca gia tri khong hop le nhu '?' va cac outlier.

Loai bo cac gia tri khong hop le va chuyen truong Inches sang dang so 

```python
# chuyen cac gia tri trong truong Inches thanh so, neu khong chuyen 
# duoc thi se chuyen thanh null
df.loc[:,'Inches']=pd.to_numeric(df['Inches'], errors='coerce')

# quan sat cac ban ghi bi null
all_null_row=df[df.isnull().any(axis=1)]
all_null_row
```

```python
# Gen gia tri inches theo 5 gia tri co tan so cao nhat
top5_inches=df['Inches'].value_counts().index[:5].to_list()
# dien tat cac cac gia tri null trong truong inches bang cach chon ngay nhien 1 trong 5 
df.loc[:,'Inches']=df['Inches'].apply(lambda x: np.random.choice(top5_inches) if pd.isnull(x) else x)
# kiem tra xem trong tuong inches con gia tri null nao khong neu khong thi tra ve false
print(df['Inches'].isnull().any())
```

Ve bieu do cho truong inches de de quan sat du lieu

```python
# ve bieu do cot 
plt.figure(figsize=(10,6))
plt.hist(df['Inches'], bins=20, color='skyblue', edgecolor= 'black')
plt.xlabel('Inches')
plt.ylabel('Frequency')
plt.title('Bieu do cot theo Inches')
plt.show()

#Ve bieu do tron
all_value=df['Inches'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(all_value, labels=all_value.index, autopct='%1.1f%%', startangle=140)
plt.title('Bieu do tron cho truong Inches')
plt.show()
```

Su dung IQR de loai bo outlier

```python
Q1=df['Inches'].quantile(0.25)
Q3=df['Inches'].quantile(0.75)
IQR= Q3-Q1
# tim ra cac muc voi IQR
lower_bound=Q1 - 1.5*IQR
upper_bound= Q3 + 1.5*IQR
print( upper_bound, '  ', lower_bound)

df.loc[:,'Inches']=df['Inches'].where((df['Inches']>=lower_bound) & (df['Inches']<=upper_bound))

all_null_row=df[df.isnull().any(axis=1)]
all_null_row
print('len =',len(all_null_row))
```

Generate cac gia tri cho cac outlier

```python
top5_inches=df['Inches'].value_counts().nlargest(5) # 5 gia tri pho bien nhat

prob=top5_inches/top5_inches.sum() # tinh xac xuat cho tung gia tri

# chon cac gia tri theo phan phoi da cho truoc
df.loc[:,'Inches']=df['Inches'].apply(lambda x: np.random.choice(top5_inches.index, p=prob) if pd.isnull(x) else x)

# ve bieu do cot 
plt.figure(figsize=(10,6))
plt.hist(df['Inches'], bins=20, color='skyblue', edgecolor= 'black')
plt.xlabel('Inches')
plt.ylabel('Frequency')
plt.title('Bieu do cot theo Inches')
plt.show()

#Ve bieu do tron
all_value=df['Inches'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(all_value, labels=all_value.index, autopct='%1.1f%%', startangle=140)
plt.title('Bieu do tron cho truong Inches')
plt.show()
```

Kiem tra cac ban ghi khong phai la so

Clean truong Weight

Loai bo 'kg' khoi truong Weight

```python
df.loc[:,'Weight']=df['Weight'].str.removesuffix('kg')
```

Chuyen truong Weight thanh so va generate gia tri cho cai ban ghi null

```python
# chuyen Truong Weight thanh so neu khong chuyen dc thi thanh null
df.loc[:,'Weight']=pd.to_numeric(df['Weight'], errors='coerce')
# Kiem tra xem co ban ghi nao khong the chuyen thanh so 
print(df['Weight'].isnull().any())

# Generate gia tri cho cac ban ghi null

#Lay top 5 gia tri co tan so cao nhat
top5_Weight=df['Weight'].value_counts().nlargest(5)
prob=top5_Weight / top5_Weight.sum() # lay xac xuat tung gia tri
df.loc[:,'Weight']=df['Weight'].apply(
    lambda x: np.random.choice(top5_Weight.index,p=prob) if pd.isnull(x) else x)

print(df['Weight'].isnull().any()) # Moi thu deu on
```

Ve bieu do de de quan sat truong Weight de hon 

```python
# Ve bieu do cot
plt.figure(figsize=(10,6))
plt.hist(df['Weight'], bins=20,color='skyblue', edgecolor='black')
plt.xlabel( 'Weight')
plt.ylabel('Frequency')
plt.title('Bieu do the hien phan bo cua Weight')
plt.show()
```

Ta thay truong Weigh van con mot so outlier, ta se su dung IQR de loai bo no roi se generate lai no.

```python
Q1=df['Weight'].quantile(0.25)
Q3=df['Weight'].quantile(0.75)
IQR= Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound, upper_bound)

df.loc[(df['Weight']<lower_bound) | (df['Weight']>upper_bound) ,'Weight']=np.nan

#Lay top 5 gia tri co tan so cao nhat
top5_Weight=df['Weight'].value_counts().nlargest(5)
prob=top5_Weight / top5_Weight.sum() # lay xac xuat tung gia tri
# generate gia tri theo phan bo 
df.loc[:,'Weight']=df['Weight'].apply(lambda x: np.random.choice(top5_Weight.index,p=prob) if pd.isnull(x) else x)
# kiem tra con gia tri null nao khong
print(df['Weight'].isnull().any()) # Moi thu deu on
```

Ve bieu do

```python
# Ve bieu do cot
plt.figure(figsize=(10,6))
plt.hist(df['Weight'], bins=20,color='skyblue', edgecolor='black')
plt.xlabel( 'Weight')
plt.ylabel('Frequency')
plt.title('Bieu do the hien phan bo cua Weight')
plt.show()

all_value=df['Weight'].value_counts()
# Ve bieu do tron
plt.figure(figsize=(8,8))
plt.pie(all_value, labels=all_value.index, autopct='%1.1f%%', startangle=90)
plt.title('Phan bo cau weight')
plt.show()
```

```python

# Đọc dữ liệu từ file CSV
file_path = './laptopData_filtered_cleaned.csv'
df2 = pd.read_csv(file_path)

# Danh sách các hãng cần giữ cho cột Company
companies = ['Lenovo', 'Dell', 'HP', 'Asus', 'MSI']

# 1. Xử lý cột Company và tạo các cột cho mỗi hãng cần giữ
for company in companies:
    df2[company] = df2['Company'].apply(lambda x: 1 if x == company else 0)

# Tạo cột Other_Comp cho những hãng không có trong danh sách
df2['Other_Comp'] = df2['Company'].apply(lambda x: 1 if x not in companies else 0)

# 2. Danh sách các giá trị cần giữ cho cột TypeName
type_name_values = ['Notebook', 'Gaming', 'Ultrabook', '2 in 1 Convertible', 'Workstation', 'Netbook']

# Xử lý cột TypeName và tạo các cột cho mỗi giá trị TypeName cần giữ
for type_name in type_name_values:
    df2[type_name] = df2['TypeName'].apply(lambda x: 1 if x == type_name else 0)

# Tạo cột Other_TypeName cho các giá trị không có trong danh sách
df2['Other_TypeName'] = df2['TypeName'].apply(lambda x: 1 if x not in type_name_values else 0)

#  3. Danh sách các giá trị cần giữ cho cột Ram
# Chuyển đổi cột 'Ram' thành số thực trong cột 'Ram_Number'
df2['Ram_Number'] = pd.to_numeric(df2['Ram'], errors='coerce')  # 'coerce' sẽ chuyển các giá trị không hợp lệ thành NaN

# 4. Xử lý cột OpSys để chia thành 'Windows' và 'Other_OpSys'
df2['Windows'] = df2['OpSys'].apply(lambda x: 1 if x in ['Windows 10', 'Windows 7'] else 0)

# Tạo cột Other_OpSys cho các giá trị không phải Windows
df2['Other_OpSys'] = df2['OpSys'].apply(lambda x: 1 if x not in ['Windows 10', 'Windows 7'] else 0)

# 5. Giữ lại chỉ các cột cần thiết (các cột đã xử lý từ 'Company', 'TypeName', 'Ram', 'OpSys')
columns_to_save = companies + ['Other_Comp'] + type_name_values + ['Other_TypeName'] + ['Ram_Number'] + ['Windows', 'Other_OpSys']

# Lấy các cột cần thiết
df_to_save = df2[columns_to_save]
```
